# ISE Work Tracker Web Application

## User Maintain

### Adding User

ขั้นตอนที่ 1
Admin กดสร้างใน Supabase

1. ไปที่ Supabase Dashboard → Authentication → Users
2. กดปุ่ม “Add user”
3. เลือกชนิด 
    - แบบ “Email + Password” → กรอก email + password ให้เค้า (เอาไปแจ้งผู้ใช้)
    - หรือแบบ “Invite via email” → ระบบส่งลิงก์ให้ user ตั้งรหัสเอง
4. กดสร้าง

ขั้นตอนที่ 2 เพิ่มข้อมูลจาก Authen ไปยัง Database

```python
insert into public.profiles (email, display_name, role, team_id)
values (
  'new.user@chula.ac.th',
  'New User',
  'user',
  (select id from public.teams where name = 'Academic' limit 1)
);
```

### Role or Team Changing

```python
# เปลี่ยน role
update public.profiles
set role = 'manager'
where email = 'someone@chula.ac.th';

# เปลี่ยน Team
update public.profiles
set team_id = (select id from public.teams where name = 'International' limit 1)
where email = 'someone@chula.ac.th';
```

### Delete user or Flag user (พักสิทธิ์)

```python
# Delete user
delete from public.profiles
where email = 'resign.user@chula.ac.th';


# Flag
update public.profiles
set is_active = false
where email = 'someone@chula.ac.th';
```
